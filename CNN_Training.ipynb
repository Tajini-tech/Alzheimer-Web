```python
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import time
import glob
from skimage import io
import nibabel as nib
from skimage.transform import resize
import random
from sklearn.model_selection import train_test_split, StratifiedKFold, KFold
from collections import Counter
import setGPU
import os
import math
import random
import time
import tensorflow
from keras.layers import Dense, Conv2D, MaxPooling2D, Dropout, Flatten
from keras.layers import BatchNormalization
from sklearn.metrics import roc_curve
from sklearn.metrics import auc
import keras
import tensorflow as tf
import matplotlib.pyplot as plt
from keras.models import Sequential,save_model,load_model

from keras.layers import Input, Dropout, Conv2DTranspose
from keras.models import Model
from keras.layers import Conv2D, MaxPooling2D, concatenate

n_folds = 3
batch_size = 15
n_epochs = 20
lr = 0.0001
src_path_AD="/content/gdrive/MyDrive/mwp_ADNI/AD_Train"
src_path_NC="/content/gdrive/MyDrive/mwp_ADNI/NC_Train"
#src_path_AD="/content/gdrive/MyDrive/colab_AD"
#src_path_NC="/content/gdrive/MyDrive/colab_NC"
with tf.device('/device:GPU:0'):
  #AD=list()
  #NC=list()
  def Dataloader(path):
    AD=list()
    for files in glob.glob(path + "/*.nii"):
                  image = nib.load(files)
                  img = image.get_fdata().astype(np.float32)
                  img = (img-img.min())/(img.max()-img.min())
                  IMG_PX_SIZE = 105
                  img = resize(img, (IMG_PX_SIZE, IMG_PX_SIZE,IMG_PX_SIZE), anti_aliasing=True)
                  AD.append(img)
    return  np.asarray(AD)
  def simpleshift(arr, num, axis, fill_value=0):
    result = np.empty_like(arr)
    if (axis==1):
      if num > 0:
          result[:, :num, :,  :] = fill_value
          result[:, num:, :,  :] = arr[:, :-num, :,  :]
      elif num < 0:
          result[:, num:, :,  :] = fill_value
          result[:, :num, :,  :] = arr[:, -num:, :,  :]
      else:
          result[:] = arr
      return result
    elif (axis==2):
      if num > 0:
          result[:, :, :num,  :] = fill_value
          result[:, :, num:,  :] = arr[:, :, :-num,  :]
      elif num < 0:
          result[:, :, num:,  :] = fill_value
          result[:, :, :num,  :] = arr[:, :, -num:,  :]
      else:
          result[:] = arr
      return result
    else:
      return None
  def get_model():
    model=Sequential()
    model.add(Conv2D(5, kernel_size=3, activation='relu', padding="same",
                     input_shape=(105, 105, 1),name="conv1"))
    model.add(BatchNormalization())
    model.add(MaxPooling2D(2, 2))

    model.add(Conv2D(5, kernel_size=3, activation='relu',padding="same",name="conv2"))
    model.add(BatchNormalization())
    model.add(MaxPooling2D(2, 2))

    model.add(Conv2D(5, kernel_size=3, activation='relu',padding="same",name="conv3"))
    model.add(BatchNormalization())
    model.add(MaxPooling2D(2, 2))


    model.add(Flatten())
    model.add(Dropout(0.1))
    model.add(Dense(64,activation="relu"))
    model.add(Dropout(0.1))
    model.add(Dense(32,activation="relu"))
    model.add(Dropout(0.1))
    model.add(Dense(2, activation='softmax'))
    print(model.summary())
    return model
  def get_data_set_axial(AD,NC):
    axial_AD=list()
    axial_NC=list()
    lb_axial=list()
    for i in range (AD.shape[0]):
      ##choose slices from 20 to 85 to avoid trainig with black empty scans
      for j in range(20,85):
        axial_AD.append(np.take(AD[i],j,2))
        lb_axial.append([1,0])
    for i in range (NC.shape[0]):
      for j in range(20,85):
        axial_NC.append(np.take(NC[i],j,2))
        lb_axial.append([0,1])

    axial_NC=np.asarray(axial_NC)
    axial_AD=np.asarray(axial_AD)
    lb_axial=np.asarray(lb_axial)
    axial_AD=np.reshape(axial_AD,(len(axial_AD),axial_AD[0].shape[0],axial_AD[0].shape[1],1))
    axial_NC=np.reshape(axial_NC,(len(axial_NC),axial_NC[0].shape[0],axial_NC[0].shape[1],1))
    return axial_AD,axial_NC,lb_axial

  def get_data_set_coronal(AD,NC):
    coronal_AD=list()
    coronal_NC=list()
    lb_coronal=list()
    for i in range (AD.shape[0]):
      for j in range(20,85):
        coronal_AD.append(np.take(AD[i],j,1))
        lb_coronal.append([1,0])
    for i in range (NC.shape[0]):
      for j in range(20,85):
        coronal_NC.append(np.take(NC[i],j,1))
        lb_coronal.append([0,1])
    coronal_NC=np.asarray(coronal_NC)
    coronal_AD=np.asarray(coronal_AD)
    lb_coronal=np.asarray(lb_coronal)
    coronal_AD=np.reshape(coronal_AD,(len(coronal_AD),coronal_AD[0].shape[0],coronal_AD[0].shape[1],1))
    coronal_NC=np.reshape(coronal_NC,(len(coronal_NC),coronal_NC[0].shape[0],coronal_NC[0].shape[1],1))
    return coronal_AD,coronal_NC,lb_coronal
  def get_data_set_sagital(AD,NC):
    sagital_AD=list()
    sagital_NC=list()
    lb_sagital=list()
    for i in range (AD.shape[0]):
      for j in range(20,85):
        sagital_AD.append(np.take(AD[i],j,0))
        lb_sagital.append([1,0])
    for i in range (NC.shape[0]):
      for j in range(20,85):
        sagital_NC.append(np.take(NC[i],j,0))
        lb_sagital.append([0,1])
    sagital_NC=np.asarray(sagital_NC)
    sagital_AD=np.asarray(sagital_AD)
    lb_sagital=np.asarray(lb_sagital)
    sagital_AD=np.reshape(sagital_AD,(len(sagital_AD),sagital_AD[0].shape[0],sagital_AD[0].shape[1],1))
    sagital_NC=np.reshape(sagital_NC,(len(sagital_NC),sagital_NC[0].shape[0],sagital_NC[0].shape[1],1))
    return sagital_AD,sagital_NC,lb_sagital


  #################################################

  def gen_data(data,label):
      while True:
          indices = list(range(len(data)))
          random.shuffle(indices)
          for i in indices:
              x = data[i]
              y = label[i]
              yield x, y
  ####################################################
  def gen_folds(data_1,data_2,label):
      indices_1 = list(range(len(data_1)))
      indices_2 = list(range(len(data_1),len(data_1)+len(data_2)))
      indices_folds = list(range(n_folds))
      folds_1 = list(StratifiedKFold(n_splits=n_folds, shuffle=True, random_state=1).split(data_1,label[:len(data_1),0]))
      folds_2 = list(StratifiedKFold(n_splits=n_folds, shuffle=True, random_state=1).split(data_2,label[len(data_1):len(data_1)+len(data_2),0]))
      for i in indices_folds:
          x_train = np.concatenate((data_1[folds_1[i][0]],data_2[folds_2[i][0]]),axis=0)
          y_train = np.concatenate((label[folds_1[i][0]],label[folds_2[i][0]+len(data_1)]),axis=0)
          x_test = np.concatenate((data_1[folds_1[i][1]],data_2[folds_2[i][1]]),axis=0)
          y_test = np.concatenate((label[folds_1[i][1]],label[folds_2[i][1]+len(data_1)]),axis=0)
          yield x_train,y_train,x_test,y_test
  ############################################################################
  def gen_batch(data,label):
      print ("in batch generator")
      counter = Counter([tuple(ele) for ele in label])
      len_1 = list(counter.items())[0][1]
      len_2 = list(counter.items())[1][1]
      #print("count elemnts ",counter.elements)
      #print(list(counter.items())[0][0],list(counter.items())[0][1],list(counter.items())[1][0],list(counter.items())[1][1])
      data_1_gen = gen_data(data[:len_1,],label[:len_1,])
      data_2_gen = gen_data(data[len_1:len_1+len_2,],label[len_1:len_1+len_2,])
      ds_1_cnt = 0
      ds_2_cnt = 0
      flag = 0
      while True:
          x_batch = []
          y_batch = []
          batch_cnt = 0
          if len_1 >= len_2:
              while batch_cnt+2 < batch_size:
                  if ds_1_cnt < len_1:
                      x_1, y_1 = next(data_1_gen)
                      x_batch.append(x_1)
                      y_batch.append(y_1)
                      ds_1_cnt += 1
                      batch_cnt += 1
                      x_2, y_2 = next(data_2_gen)
                      x_batch.append(x_2)
                      y_batch.append(y_2)
                      ds_2_cnt += 1
                      batch_cnt += 1
                  else:
                      break
              #print("<<<",ds_1_cnt,len_1)
              if ds_1_cnt < len_1:
                  #print("<<<",batch_cnt,batch_size)
                  if batch_cnt < batch_size:
                      if flag == 0:
                          x_2, y_2 = next(data_2_gen)
                          x_batch.append(x_2)
                          y_batch.append(y_2)
                          ds_2_cnt += 1
                          batch_cnt += 1
                          flag = 1
                      else:
                          x_1, y_1 = next(data_1_gen)
                          x_batch.append(x_1)
                          y_batch.append(y_1)
                          ds_1_cnt += 1
                          batch_cnt += 1
                          flag = 0
                  batch_cnt = 0
              else:
                  #print("$$$",ds_1_cnt,ds_2_cnt)
                  if batch_cnt != 0:
                      batch_cnt = 0
                      yield np.array(x_batch),np.asarray(y_batch)
                  break
          else:
              while batch_cnt+2 < batch_size:
                  if ds_2_cnt < len_2:
                      x_2, y_2 = next(data_2_gen)
                      x_batch.append(x_2)
                      y_batch.append(y_2)
                      ds_2_cnt += 1
                      batch_cnt += 1
                      x_1, y_1 = next(data_1_gen)
                      x_batch.append(x_1)
                      y_batch.append(y_1)
                      ds_1_cnt += 1
                      batch_cnt += 1
                  else:
                      break
              #print("<<< ",ds_2_cnt,len_2)
              if ds_2_cnt < len_2:
                  if batch_cnt < batch_size:
                      if flag == 0:
                          x_1, y_1 = next(data_1_gen)
                          x_batch.append(x_1)
                          y_batch.append(y_1)
                          ds_1_cnt += 1
                          batch_cnt += 1
                          flag = 1
                      else:
                          x_2, y_2 = next(data_2_gen)
                          x_batch.append(x_2)
                          y_batch.append(y_2)
                          ds_2_cnt += 1
                          batch_cnt += 1
                          flag = 0
                      batch_cnt = 0
              else:
                  #print("$$$",ds_1_cnt,ds_2_cnt)
                  if batch_cnt != 0:
                      batch_cnt = 0
                      yield np.array(x_batch),np.asarray(y_batch)
                  break

          #print(ds_1_cnt,ds_2_cnt)
          yield np.array(x_batch),np.asarray(y_batch)
  ###########################################################
  def one_vs_one_train(ds_name_1,ds_name_2):
      AD=Dataloader(src_path_AD)
      NC=Dataloader(src_path_NC)
      ds_11, ds_22, lb12 = get_data_set_axial(AD,NC)
      del AD
      del NC
      ####

      ds_1 = np.concatenate((ds_11,simpleshift(ds_11, -2, 1)),axis=0)
      ds_2 = np.concatenate((ds_22,simpleshift(ds_22, -2, 1)),axis=0)
      #ds_1 = np.concatenate((ds_11,np.flip(ds_11, axis=2)),axis=0) # L/R flip
      #ds_2 = np.concatenate((ds_22,np.flip(ds_22, axis=2)),axis=0)

      lb = np.concatenate((lb12[:9750],lb12[:9750],lb12[9750:9750+13000],lb12[9750:9750+13000]),axis=0)
      #print("shape of lb after",lb.shape)

      #####
      folds_gen = gen_folds(ds_1,ds_2,lb)
      fold_index = 0
      for x_train,y_train,x_test,y_test in folds_gen:
        fold_index += 1
        model = get_model()
        opt = tensorflow.keras.optimizers.Adam(lr)
        model.compile(loss='categorical_crossentropy',optimizer=opt, metrics=['accuracy'])
        for i in range(n_epochs):
              print(" epoch : ",i)
              batch_gen = gen_batch(x_train,y_train)
              for x_batch, y_batch in batch_gen:
                  history_train = model.fit(x_batch, y_batch, epochs=1, verbose=0, shuffle=True,)
              training_history = '/content/gdrive/MyDrive/Freitag_model/Freitag_scores/axial_raw/training_history_{}_vs_{}_fold_index_{}.txt'.format(str(ds_name_1),str(ds_name_2),str(fold_index))
              with open(training_history,"a+") as f:
                  f.write('{},{},{}\n'.format(str(i), str(history_train.history['loss'][0]), str(history_train.history['accuracy'][0])))
              history_eval = model.evaluate(x_test,y_test)
              print(" eval_acc :",history_eval[1])
              print("eval_loss : ",history_eval[0])
              #save model at the end of each epoch
              model.save("/content/gdrive/MyDrive/Freitag_model/Freitag_weights/axial_raw/ ep % d in fo % d eval_acc %f eval_loss %f.h5" %(i, fold_index ,history_eval[1] ,history_eval[0]))
              evaluation_history = '/content/gdrive/MyDrive/Freitag_model/Freitag_scores/axial_raw/evaluation_history_{}_vs_{}_fold_index_{}.txt'.format(str(ds_name_1),str(ds_name_2),str(fold_index))
              with open(evaluation_history,"a+") as f:
                  f.write('{},{},{}\n'.format(str(i), str(history_eval[1]),str(history_eval[0])))
              y_pred_keras = model.predict(x_test)
              fpr_keras, tpr_keras, thresholds_keras = roc_curve(y_test[:,0], y_pred_keras[:,0])
              auc_keras = auc(fpr_keras, tpr_keras)
              ##plot roc curve
              plt.plot(fpr_keras,tpr_keras,label="AUC="+str(auc_keras),color="darkorange")
              plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
              plt.ylabel('True Positive Rate')
              plt.xlabel('False Positive Rate')
              plt.legend(loc=4)
              plt.show()
              ROC_AUC_history = '/content/gdrive/MyDrive/Freitag_model/Freitag_scores/axial_raw/ROC_AUC_history_{}_vs_{}_fold_index_{}.txt'.format(str(ds_name_1),str(ds_name_2),str(fold_index))
              with open(ROC_AUC_history,"a+") as f:
                  f.write('{},{}\n'.format(str(i), auc_keras))

  one_vs_one_train("AD","NC")
```
